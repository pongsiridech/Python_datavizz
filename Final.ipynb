````python
import pandas as pd
from matplotlib import pyplot as plt
````

# ตรวจสอบ missing

````python
Covid.isnull().any()
````

# แทนค่า missing ด้วย unknown

````python
Covid['Province/State'] = Covid['Province/State'].fillna('unknown')
````

````python
Covid.isnull().any()
````

````python
Covid
````

# จัดการตาราง
### แยกคอลัมน์ DATE เอาวันที่และเวลาออกจากกัน 

````python
D = []
for i in range(Covid.shape[0]):
    A = Covid.iloc[i]['Date'].split(' ')
    A = A[0].replace('12:00:00', ' ')
    D.append(A)
D
````

## เพิ่มคอลัมน์ Date_new ในตาราง Covid

````python
Covid['Date_new'] = D
````

## ลบคอลัมน์ Date กับ Last Update ในตาราง Covid

````python
del Covid['Date']
del Covid['Last Update']
````

````python
Covid = Covid.reset_index(drop=True)
````

````python
Covid
````

## สร้างตารางใหม่ที่มีเฉพาะข้อมูลของสหรัฐอเมริกาและออสเตเลีย เพื่อที่จะนำไปสร้างการนำเสนอข้อมูลใน power bi

````python
US
````

````python
AU = AU.reset_index(drop=True)
````

````python
AU
````

# รวมตาราง US กับ AU

````python
df =  [US,AU]
US_AU = pd.concat(df)
````

````python
US_AU
````

````python
US_AU = US_AU.reset_index(drop=True)
````

````python
US_AU
````

# สร้างกราฟ

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
x = Covid.groupby(['Date_new'], as_index=False).sum()['Date_new']
y = Covid.groupby(['Date_new'], as_index=False).sum()['Confirmed']
plt.figure(figsize=[20,10],dpi=100)
plt.gca(facecolor='oldlace')
plt.plot(x,y,'R',label = "Confirmed",marker='o')
plt.legend()
plt.xticks(rotation = 45)
plt.xlabel('วันที่',fontname='Tahoma',fontsize=25)
plt.ylabel('จำนวน(คน)',fontname='Tahoma',fontsize=25)
plt.title('กราฟแสดงจำนวนผู้ติดเชื้อ Covid-19',fontname='Tahoma',fontsize=20)
````

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
x = Covid.groupby(['Country'], as_index=False).sum()['Country']
y = Covid.groupby(['Country'], as_index=False).sum()['Confirmed']
plt.figure(figsize=[20,10],dpi=100)
plt.gca(facecolor='oldlace')
plt.plot(x,y,'G',label = "Confirmed",marker='o')
plt.legend()
plt.xticks(rotation = 45)
plt.xlabel('ประเทศ',fontname='Tahoma',fontsize=25)
plt.ylabel('จำนวน(คน)',fontname='Tahoma',fontsize=25)
plt.title('กราฟแสดงจำนวนผู้ติดเชื้อ Covid-19 ในแต่ละประเทศ',fontname='Tahoma',fontsize=20)
````

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
x = Covid[Covid['Country']=='Thailand']['Date_new']
y1 = Covid[Covid['Country']=='Thailand']['Confirmed']
y2 = Covid[Covid['Country']=='Thailand']['Deaths']
y3 = Covid[Covid['Country']=='Thailand']['Recovered']
plt.figure(figsize=[10,5],dpi=80)
ax = plt.gca(facecolor='#EFEEFE')
plt.fill_between( x, y1, color="r",alpha=0.2,label = "Confirmed")
plt.plot(x, y1, color="Slateblue", alpha=0.6)
plt.fill_between( x, y2, color="g",alpha=0.2,label = "Deaths")
plt.plot(x, y2, color="plum", alpha=0.6)
plt.fill_between( x, y3, color="orange",alpha=0.2,label = "Recovered")
plt.plot(x, y3, color="turquoise", alpha=0.6)
plt.legend()
plt.xticks(rotation = 90)
ax.yaxis.set_ticks_position('right')
plt.xlabel('วันที่',fontname='Tahoma')
plt.ylabel('จำนวน(คน)',fontname='Tahoma')
plt.title('กราฟแสดงการเปรียบเทียบจำนวนผู้ติดเชื้อ,ผู้ที่ได้รับการรักษา,ผู้ที่เสียชีวิตในโรค Covid-19 ในประเทศไทย เดือน ม.ค.- ก.พ. 2563',fontname='Tahoma',fontsize=12)
````
