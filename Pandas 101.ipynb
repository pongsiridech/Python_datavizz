### delete row with missing value
### เป็นตัวอย่างแสดงการลบ missing ใน คอลัมน์ zipcode โดยใช้คำสั่ง .dropna(subset = ['zipcode'])

### เป็นตัวอย่างแสดงการลบ missing ใน คอลัมน์ zipcode, number_of_chains โดยใช้คำสั่ง .dropna(subset = ['zipcode','number_of_chains'])

```python
w_restaurant_df.dropna(subset = ['zipcode','number_of_chains'])
```

# อัพเดทตาราง w_restaurant_df เมื่อแก้คอลัมน์เสร็จ

```python
w_restaurant_df = w_restaurant_df.dropna(subset = ['zipcode','number_of_chains'])
```

## fill default value to the data
###  เติม 0 ใส่ missing ของ parking

```python
w_restaurant_df['parking'].fillna(0)
```

### เป็นการบันทึก คอลัมน์ parking ที่ทำการแก้ใหม่แล้ว ในคอลัมน์ parking เดิม

```python
w_restaurant_df['parking'] = w_restaurant_df['parking'].fillna(0)
```
