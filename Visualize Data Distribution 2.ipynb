```python
Debt_df[['Region and province',2549]] #เลือก columns ที่เราสนใจ
```

```python
Debt_df_debt = Debt_df[['Region and province',2549]]
```

```python
Debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis=0,inplace=True) #ลบแถวที่ไม่เกี่ยวข้องหรือไม่สนใจทิ้งไป
```

```python
Debt_df_debt.head()
```

```python
import folium #คำสั่ง map
```

```python

import pandas as pd
```

```python
data_to_plot = Debt_df_debt
```

```python
data_to_plot.head()
```

```python
country_geo = 'thai.json'  #ดาวน์โหลดข้อมูลเชิงภูมิศาสตร์ของประเทศไทย
```

```python
new_col = list()   #ลบช่องว่างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์เชื่อมกันได้
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province']=new_col
```

```python
map = folium.Map(location=[10,100], zoom_start=5) #กำหนดพิกัด ลองติจูด และละติจูดของประเทศไทย
```

```python
map
```

```python
map.choropleth(geo_data=country_geo,data=data_to_plot,
              columns=['Region and province',2549],
              key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
              legend_name='Average Debt')
```

```python
map.save('plot_data.html') #บันทึกไฟล์เป็นแบบเว็บไซต์
```
